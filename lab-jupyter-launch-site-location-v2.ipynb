## Task 1: Mark all launch sites on a map


First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site. 


```python
# Download and read the `spacex_launch_geo.csv`
spacex_csv_file = wget.download('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.


We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.


```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example, 


```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

and you should find a small yellow circle near the city of Houston and you can zoom-in to see a larger circle. 


Now, let's add a circle for each launch site in data frame `launch_sites`


_TODO:_  Create and add `folium.Circle` and `folium.Marker` for each launch site on the site map


An example of folium.Circle:

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5.5)
# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label

# Loop over the DataFrame rows to add markers
for _, row in launch_sites_df.iterrows():
    coords = [row['Lat'], row['Long']]
    name = row['Launch Site']
    
    # Add a circle marker
    circle = folium.Circle(
        location=coords,
        radius=1000,  # Adjust radius as needed
        color='#d35400',  # Circle color
        fill=True,
        fill_color='#ffa07a'
    ).add_child(folium.Popup(name))
    site_map.add_child(circle)
    
    # Add a marker with a text label
    marker = folium.Marker(
        location=coords,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{name}</b></div>'
        )
    )
    site_map.add_child(marker)

# Save the map to an HTML file or display it
site_map.save("launch_sites_map_with_labels.html")
site_map
```

# Task 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records. 
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

_TODO:_ Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

_TODO:_ For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
from folium.plugins import MarkerCluster
from folium import Icon, Popup

# Add a MarkerCluster to the site map
marker_cluster = MarkerCluster().add_to(site_map)

# Loop through the spacex_df DataFrame to add markers
for _, row in spacex_df.iterrows():
    coords = [row['Lat'], row['Long']]
    
    # Create a Marker object
    marker = folium.Marker(
        location=coords,
        icon=Icon(color='white', icon_color=row['marker_color']),  # Use the 'marker_color' for icon
        popup=Popup(
            f"<b>Launch Site:</b> {row['Launch Site']}<br><b>Class:</b> {'Success' if row['class'] == 1 else 'Failure'}",
            max_width=300
        )
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Save or display the updated map
site_map.save("launch_sites_map_with_clusters.html")
site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition

# Add Mouse Position to the map to show Lat, Long dynamically
mouse_position = MousePosition(
    position='topleft',         # Position widget in the top-left corner
    separator=' Long: ',        # Separator between Lat and Long
    empty_string='No data',     # Text shown when no position is available
    lng_first=False,            # Display Latitude before Longitude
    num_digits=5,               # Show coordinates with 5 decimal places
    prefix='Lat:',              # Prefix for Latitude
)

# Add MousePosition to the site map
site_map.add_child(mouse_position)

# Save or display the updated map
site_map.save("map_with_mouse_position.html")
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

_TODO:_ After obtained its coordinate, create a `folium.Marker` to show the distance


```python
from folium.features import DivIcon
from geopy.distance import geodesic

# Coordinates for the launch site and the closest coastline point
launch_site_coords = (28.57261, -80.57163)  # Example coordinates (adjust accordingly)
coastline_coords = (28.573255, -80.646895)  # Example coastline coordinates (adjust accordingly)

# Calculate the distance between the launch site and the coastline point
distance_km = geodesic(launch_site_coords, coastline_coords).km

# Create a folium map centered at the launch site
map = folium.Map(location=launch_site_coords, zoom_start=15)

# Add a marker for the launch site
folium.Marker(
    location=launch_site_coords,
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(map)

# Add a marker for the coastline point with distance information
distance_marker = folium.Marker(
    location=coastline_coords,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color: #d35400;"><b>{:10.2f} KM</b></div>'.format(distance_km),
    )
)

distance_marker.add_to(map)

# Draw a line between the launch site and the coastline point
folium.PolyLine([launch_site_coords, coastline_coords], color='green', weight=2.5, opacity=1).add_to(map)

# Display the map
map.save("map_with_distance_marker.html")
map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Coordinates for the launch site and closest features
launch_site_coords = (28.57261, -80.57163)   # Example coordinates (adjust accordingly)
closest_city_coords = (28.07774, -80.61525)  # Example city coordinates (adjust accordingly)
closest_railway_coords = (28.57108, -80.58541)  # Example railway coordinates (adjust accordingly)
closest_highway_coords = (28.56104, -80.57043)  # Example highway coordinates (adjust accordingly)

# Calculate distances to the closest features
distance_to_city_km = geodesic(launch_site_coords, closest_city_coords).km
distance_to_railway_km = geodesic(launch_site_coords, closest_railway_coords).km
distance_to_highway_km = geodesic(launch_site_coords, closest_highway_coords).km

# Create a folium map centered at the launch site
map = folium.Map(location=launch_site_coords, zoom_start=15)

# Add a marker for the launch site
folium.Marker(
    location=launch_site_coords,
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(map)

# Add markers for the closest features with distance information
city_marker = folium.Marker(
    location=closest_city_coords,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color: #d35400;"><b>{:10.2f} KM</b></div>'.format(distance_to_city_km),
    )
)
city_marker.add_to(map)

railway_marker = folium.Marker(
    location=closest_railway_coords,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color: #d35400;"><b>{:10.2f} KM</b></div>'.format(distance_to_railway_km),
    )
)
railway_marker.add_to(map)

highway_marker = folium.Marker(
    location=closest_highway_coords,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color: #d35400;"><b>{:10.2f} KM</b></div>'.format(distance_to_highway_km),
    )
)
highway_marker.add_to(map)

# Draw lines between the launch site and the closest features
folium.PolyLine([launch_site_coords, closest_city_coords], color='red', weight=2.5, opacity=1).add_to(map)
folium.PolyLine([launch_site_coords, closest_railway_coords], color='green', weight=2.5, opacity=1).add_to(map)
folium.PolyLine([launch_site_coords, closest_highway_coords], color='blue', weight=2.5, opacity=1).add_to(map)

# Save the map
map.save("map_with_features_distance.html")
map
```
